### Binomial distribution

```python
# a bank issues credit card to the customers under the schemes of master card based on passed data it is known that 60% of te customers pay the bill on time.if a sample of 7 accounts is selected at random current databases , construct the binomial probability distribution of account 
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n=7
```

```python
p=0.6
```

```python
k=np.arange(0,8)
k
```

```python
binomial = stats.binom.pmf(k,n,p)
```

```python
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.title('binomial: n=%i , p=%0.2f' % (n,p), fontsize=15)
plt.xlabel('number of success')
plt.ylabel('probability of success')
plt.show()
```
